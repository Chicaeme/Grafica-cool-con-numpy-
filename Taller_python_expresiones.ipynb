```python
# valor de x que es indepediente y es un arreglo

X = np.linspace(0, 10, 100)

X
```

```python
Y = np.log(X + np.sin(np.pi / 2)) + np.sqrt(np.arctan(X/5 + np.pi/8))
Y
```

```python
import matplotlib.pyplot as plt #para comenzar a graficar
```

```python
migrafica = plt.figure(figsize=(10, 5))

plt.plot(X, Y)
plt.plot(X, Y, marker='o', linestyle='-', color='b', linewidth=2, markersize=8, label='Estaturas')

# Añadir etiquetas y título con estilo
plt.xlabel('Estudiantes', fontsize=12, fontweight='bold', color='gray')
plt.ylabel('Estatura (cm)', fontsize=12, fontweight='bold', color='gray')
plt.title('Estaturas de los estudiantes', fontsize=16, fontweight='bold')

# Añadir cuadrícula y leyenda
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper left', fontsize=10)


# Añadir un fondo de color al gráfico
plt.gca().set_facecolor('#f7f7f7')

# Añadir una línea horizontal en la media de estaturas
media_estaturas = sum(Y) / len(Y)
plt.axhline(y=media_estaturas, color='r', linestyle='--', label='Media de estaturas')

# Mostrar el gráfico con más detalles estéticos
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Definir datos ficticios para X, Y y Z (estudiantes, estatura y edad)
estudiantes = ['Estudiante 1', 'Estudiante 2', 'Estudiante 3', 'Estudiante 4', 'Estudiante 5']
estatura = [160, 165, 170, 175, 180]  # Estaturas ficticias en cm
edad = [18, 19, 20, 21, 22]  # Edades ficticias en años

# Asignar valores numéricos a los estudiantes
indices = np.arange(len(estudiantes))

# Crear la figura para el gráfico 3D
fig = plt.figure(figsize=(30, 10))
ax = fig.add_subplot(111, projection='3d')

# Trazar el gráfico de dispersión tridimensional
ax.scatter(indices, edad, estatura, color='blue', s=100, label='Estaturas')

# Añadir etiquetas y título
ax.set_xticks(indices)
ax.set_xticklabels(estudiantes)
ax.set_xlabel('Estudiantes', fontsize=12)
ax.set_ylabel('Edad', fontsize=12)
ax.set_zlabel('Estatura (cm)', fontsize=12)
ax.set_title('Estaturas de los estudiantes', fontsize=16)

# Añadir una leyenda
ax.legend()

# Rotar los ejes para una mejor visualización
ax.view_init(elev=20, azim=45)

# Mostrar el gráfico 3D
plt.show()
```
